```python
import pandas as pd
import numpy as np
from pandas.api.types import is_numeric_dtype
import pydotplus 
from IPython.display import Image
import sklearn
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer

df = pd.read_csv('chess.csv')
df = df[df['rated'] != False] #Removes unrated games
df['opening'] = df['opening_name'].apply(lambda x: ' '.join(x.split(' ')[:2]))
df['opening'] = df['opening'].apply(lambda x: x[:-1] if str(x).endswith(':') else x) #Limits opening name to two words to reduce all the variations
```

```python
documents = df['moves'].values.astype("U") #looks for all unicde
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
features = vectorizer.fit_transform(documents) 
```

```python
k = 20
model = KMeans(n_clusters = k, init='k-means++',max_iter=200, n_init = 1)
model.fit(features)
```

```python
df['cluster'] = model.labels_
print(df.head(20))
```

```python
print("Cluster Centroids: \n")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()

for i in range(k):
  print('Cluster %d:' % i)
  for j in order_centroids[i, :10]:
    print(' %s' % terms[j])
  print('------------')
```
